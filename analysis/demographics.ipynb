```python
import glob
import pandas as pd

csv_files = sorted(glob.glob("temp/demographic-data/*.csv"))
df = pd.concat((pd.read_csv(f) for f in csv_files), ignore_index=True)
```

```python
df = df[df["Status"].isin(["APPROVED", "AWAITING REVIEW"])]
```

```python
age_numeric = pd.to_numeric(df["Age"], errors = "coerce")
age_stats = age_numeric.agg(["mean", "std"])

sex_percent = (
    df["Sex"]
        .value_counts(dropna = False, normalize = True)
        .mul(100)
        .rename("percent")
)

ethnicity_counts = df["Ethnicity simplified"].value_counts(dropna = False, normalize = True).mul(100).rename("percent")

age_stats, sex_percent, ethnicity_counts
```
